```python
import csv
myList = []
with open('c:\\Users\\lenovo\\Desktop\\iris.data') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
         if len(row) == 5:
             myList.append([float(row[0]), float(row[1]), float(row[2]), float(row[3])])

print(myList)
```

```python
import numpy as np
npArray = np.array(myList)
for i in range(0,len(npArray[0])):
    column_min = np.min(npArray[:,i])
    
    
    
    column_max = np.max(npArray[:, i])
    
    
    
    column_mediana = np.median(npArray[:, i])
    
    
    
    
    column_average = np.average(npArray[:, i])
    
    
    
    
    column_norm = (npArray[:,i]-min(npArray[:,i])) / (max(npArray[:,i])-min(npArray[:,i]))
    print('Minimul de pe coloana:', i ,'este',column_min)
    print('Maximul de pe coloana:', i ,'este',column_max)
    print('Mediana de pe coloana:', i , 'este',column_mediana)
    print('Media de pe coloana:', i , 'este',column_average)
    print('Valorile normale de pe coloana:', i , 'este',column_norm)
    
    
    
```
