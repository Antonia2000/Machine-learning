```python
import numpy as np

w = [[0, 1, 0],[],[]]
x =np.array( [[2, 1, -1],[0, -1, -1]])

d = [-1,1]        #desired
c = 0.1
net = [[]]*2   
f_net = []        #vector cu net
s_net = []        #vector cu semnul netului

def perceptron(x,w,d,c, nr_iter):
    for j in range(nr_iter):
        for i in range(len(x)):
            net[i] = np.sum(np.multiply(w[i],x[i]))
            f_net.append(net)
            print("Net: ",net[i])
            s_net.append(np.sign(net[i]))
            print("semn de net:",s_net[i])
            if(d[i] != s_net[i]):
                w[i+1]=w[i]+c*2*d[i]*x[i]
                print("Pondere corectata:", w[i+1])
                print()
            else:
                w[i+1]=w[i]
                print("Pondere necorectata: ",w[i+1])
                print()
        if d == s_net:
               break
        else:
               f_net.clear()
        w[0]=w[2]
        
            
print("PERCEPTRON!")
perceptron(x,w,d,c,6)
```
